# Inference of Model
Notebook which steps through the steps for running inference. 

0. Import libraries  
1. Transform the audio file in stereo .wav format and split into 3 second clips 
2. Create mel spectrogram and save the figures
3. Run the inference on the mel spectrogram
4. Compare prediction to true value

### Step 0: Import the required packages and define variables

```python
import boto3
import librosa
import librosa.display
from pydub import AudioSegment
```

```python
import botocore
import json
import requests
import time
import os.path
import sys
import getopt
import argparse
import numpy as np
from pathlib import Path
import matplotlib.pyplot as plt
from datetime import datetime
```

### Step 1: Process the audio recording 


```python
def mono_to_stereo(AudioFile):

    left_channel = AudioFile
    right_channel = AudioFile

    stereo_sound = AudioSegment.from_mono_audiosegments(left_channel, right_channel)
    return stereo_sound
```

### Step 2: Create the Mel Spectrogram

```python
def make_melspgram(audio_segment):
    clip, sample_rate = librosa.load(audio_segment, sr=None)

    n_mels = 64
    n_fft = 1024
    fmin = 0
    fmax = 22050 
    hop_length = 512
    mel_spec = librosa.feature.melspectrogram(clip, n_fft=n_fft, hop_length=hop_length,
                                              n_mels=n_mels, sr=sample_rate, power=1.0,
                                              fmin=fmin, fmax=fmax)
    mel_spec_db = librosa.amplitude_to_db(mel_spec, ref=np.max)
    plt.figure(figsize=(4, 4))
    librosa.display.specshow(mel_spec_db, x_axis=None,  y_axis=None,
                             sr=sample_rate, hop_length=hop_length,
                             fmin=fmin, fmax=fmax)
    path_to_melsgplt='../data/demo/melsgplt.png'
    plt.savefig(path_to_melsgplt,bbox_inches='tight',pad_inches=0)
    #plt.show()
    plt.close('all')
    return path_to_melsgplt
```

```python
make_melspgram('../data/clips/eastern-whipbird/6316ecca-4b96-4477-8da7-e764429fa389_2000_5000.wav')
```

### Step 3: Run inference 

```python
inference_categories = ['eastern-whipbird', 'kookaburra','willie-wagtail']
client = boto3.client('sagemaker-runtime', region_name='ap-southeast-2')
endpoint_name = 'birdsong-no-tl-mel-e30-mb8-layers152-ep--2022-08-02-04-10-40'
```

```python
def run_inference(file_name):
    with open(file_name, 'rb') as f:
        payload = f.read()
        payload = bytearray(payload)
    response = client.invoke_endpoint(EndpointName=endpoint_name,
                                   ContentType='application/x-image',
                                   Body=payload)
    result = response['Body'].read()
    result = json.loads(result)
    index = np.argmax(result)
    confidence = result[index]
    #print(result)
    return inference_categories[index], confidence
```

```python
run_inference('../data/demo/melsgplt.png')
```

### Step 4: Bringing it all together 

```python
true = ['eastern-whipbird', 'eastern-whipbird','eastern-whipbird','eastern-whipbird','eastern-whipbird','eastern-whipbird','kookaburra','kookaburra', 'kookaburra','kookaburra','kookaburra',
       'kookaburra', 'willie-wagtail','willie-wagtail','willie-wagtail','willie-wagtail']
```

```python
def process_audio(audio_file):
    predictions = {}
    length = audio_file.duration_seconds
    j=0 
    for i in range(0, (int(length)-3)*1000, 500):
        json_entry = {}
        start = i
        end = i+3000
        audio_segment = audio_file[start:end]
        filename1 = '../data/demo/clip'+'_'+str(start)+'_'+str(end)+'.wav'
        audio_segment.export(filename1, format="wav")
        path_to_image = make_melspgram(filename1)
        prediction, confidence = run_inference(path_to_image)
        print('Start:{}, End:{}, \n True: {}, Prediction: {}, Confidence: {}'.format(start, end,true[j], prediction, confidence))
        j +=1

```
